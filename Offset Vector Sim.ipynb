```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#No Offset
rabi = 2*np.pi*6e3

t = np.linspace(0,400e-6,1000)
rho_x = -np.sin(rabi*t)
rho_y = 0*t
rho_z = np.cos(rabi*t)

rho = np.array([rho_x, rho_y, rho_z])

final_rho_x = rho_x[-1]
final_rho_y = rho_y[-1]
final_rho_z = rho_z[-1]
```

```python
# Plotting the vector in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the trajectory of the vector over time and add label for the legend
line, = ax.plot(rho_x, rho_y, rho_z, label='Trajectory of rho')

# # Optionally: add arrows to show the vector at different points in time
# for i in range(0, len(t), 100):  # Plot arrows at some steps in time
#     ax.quiver(0, 0, 0, rho_x[i], rho_y[i], rho_z[i], color='r', alpha=0.2)

# Add the Bloch sphere (transparent sphere with radius 1)
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the sphere, make it transparent
ax.plot_surface(x, y, z, color='b', alpha=0.1, rstride=5, cstride=5, label='Bloch Sphere')

# Add axes inside the Bloch sphere
ax.quiver(0, 0, 0, 1, 0, 0, color='k', linewidth=1, label='X-axis')
ax.quiver(0, 0, 0, 0, 1, 0, color='k', linewidth=1, label='Y-axis')
ax.quiver(0, 0, 0, 0, 0, 1, color='k', linewidth=1, label='Z-axis')

ax.quiver(0, 0, 0, final_rho_x, final_rho_y, final_rho_z, color='r', linewidth=2, label='Final rho')

ax.view_init(elev=20, azim=30)  # Adjust the elevation and azimuth

# Labels and title
ax.set_xlabel('I_x')
ax.set_ylabel('I_y')
ax.set_zlabel('I_z')
ax.set_title('3D Vector Evolution on Bloch Sphere')

# Set the aspect ratio to be equal (optional)
ax.set_box_aspect([1, 1, 1])

# # Show the legend
# ax.legend()

# Show the plot
plt.show()

# Plot all three components on the same plot
plt.figure(figsize=(10, 6))

# Plot rho_x, rho_y, and rho_z
plt.plot(t, rho_x, label='rho_x', color='r')
plt.plot(t, rho_y, label='rho_y', color='g')
plt.plot(t, rho_z, label='rho_z', color='b')

# Add labels and title
plt.title('Components of rho vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Component Values')
plt.legend()

# Show the plot
plt.show()
```

```python
#With offset
offset = 2*np.pi*5e3
rabi = 2*np.pi*6.5e3

w_e = np.sqrt(offset**2 + rabi**2)
theta = np.arctan(rabi/offset)

t = np.linspace(0,1.7e-6,1000)
rho_x = np.sin(theta)*np.sin(w_e*t)
rho_y = np.sin(theta)*np.cos(theta)*(1-np.cos(w_e*t))
rho_z = (np.cos(theta))**2 + ((np.sin(theta))**2 * np.cos(w_e*t))

rho = np.array([rho_x, rho_y, rho_z])

final_rho_x = rho_x[-1]
final_rho_y = rho_y[-1]
final_rho_z = rho_z[-1]

print(np.arctan(rho_x[-1] / rho_z[-1]) * 180/np.pi)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.transform import Rotation as R

# Parameters for cone
theta_cone = np.radians(25)     # cone angle from axis
omega = 2 * np.pi               # angular velocity
T = 1                           # total time
N = 500                         # number of points
t = np.linspace(0, T, N)
phi = omega * t                 # azimuthal angle

# Vector in local frame (cone around z-axis)
x = np.sin(theta_cone) * np.cos(phi)
y = np.sin(theta_cone) * np.sin(phi)
z = np.full_like(phi, np.cos(theta_cone))
v = np.vstack([x, y, z])

# Rotate cone axis to point along x-axis instead of z-axis
# Rotate 90 degrees around y-axis
rotation = R.from_euler('y', -80, degrees=True).as_matrix()
v_rot = rotation @ v

# Extract rotated components
rho_x, rho_y, rho_z = v_rot[0], v_rot[1], v_rot[2]
final_rho_x, final_rho_y, final_rho_z = rho_x[-1], rho_y[-1], rho_z[-1]

# Plotting the vector in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the trajectory of the vector over time
ax.plot(rho_x, rho_y, rho_z, color='r', label='Trajectory of ')

# # Optional: add arrows at intervals
# for i in range(0, len(t), 200):
#     ax.quiver(0, 0, 0, rho_x[i], rho_y[i], rho_z[i], color='r', alpha=0.5)

# Bloch sphere surface
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_s = np.outer(np.cos(u), np.sin(v))
y_s = np.outer(np.sin(u), np.sin(v))
z_s = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x_s, y_s, z_s, color='b', alpha=0.1, rstride=5, cstride=5)

# Axes
ax.quiver(0, 0, 0, 1, 0, 0, color='k', linewidth=1)
ax.quiver(0, 0, 0, 0, 1, 0, color='k', linewidth=1)
ax.quiver(0, 0, 0, 0, 0, 1, color='k', linewidth=1)

# Final vector (same color as path)
ax.quiver(0, 0, 0, final_rho_x, final_rho_y, final_rho_z, color='b', alpha = 0.5, linewidth=2)

# View angle
ax.view_init(elev=20, azim=40)

# Labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Vector Evolution on Bloch Sphere')

# Appearance
ax.grid(False)
ax.set_xticks([-1, 0, 1])
ax.set_yticks([-1, 0, 1])
ax.set_zticks([-1, 0, 1])
ax.set_box_aspect([1, 1, 1])
ax.legend()

plt.tight_layout()
plt.show()

# fig.savefig("C:\\Users\\Suber\\Desktop\\BlochSphere.pdf", format='pdf', dpi=600)
```
