```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kb = 1.38e-23
D0 = 2877.38
c1 = -54.91
delta1 = 58.73 * 1.6021773e-22 # Joules
c2 = -249.6
delta2 = 145.5 * 1.6021773e-22 # Joules

T = np.linspace(1,400,1000)

n1 = (1) / (np.exp((delta1) / (kb*T)) - 1)
n2 = (1) / (np.exp((delta2) / (kb*T)) - 1)
D = D0 + c1*n1 + c2*n2

plt.figure()
plt.plot(T,D)
plt.show()
```

```python
i100 = np.argmin(np.abs(T - 100)) #index corresponding to T=100
D_100 = D[i100]
target = D_100 - 2.5 #Find where temp where D has shifted by 2.5
idx = np.argmin(np.abs(D - target))
print(f"max temp change at 100K is: {T[idx] - T[i100]}")
```

```python
i300 = np.argmin(np.abs(T - 300)) #index corresponding to T=300
D_300 = D[i300]
target = D_300 - 5 #Find where temp where D has shifted by 5
idx = np.argmin(np.abs(D - target))
print(f"max temp change at 300K is: {T[idx] - T[i300]}")
```
